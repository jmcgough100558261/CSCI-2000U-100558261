```python
#Jesse McGough
#100558261s

#Importing everything that's needed
from matplotlib import pyplot as plt
from matplotlib import patches as pch
import calendar
import numpy as np
import os
%matplotlib inline

#Imports the files and puts them into an array
files = os.listdir('.')
temp_files = []
for file in files:
    if "temperature_2012.dat" in file:
        temp_files.append(file)

#Adds the files to the plot, giving them unique colours
plt.plotfile(temp_files[0], color='red', linewidth='2', names=calendar.month_abbr[1:13])
plt.plotfile(temp_files[1], color='pink', linewidth='2', names=calendar.month_abbr[1:13])
plt.plotfile(temp_files[2], color='blue', linewidth='2', names=calendar.month_abbr[1:13])

#Organizes and labels the plot
plt.xticks(np.arange(0,12), calendar.month_abbr[1:13])
line1=pch.Patch(color='red')
line2=pch.Patch(color='pink')
line3=pch.Patch(color='blue')
plt.legend((line1, line2, line3), ('Montreal', 'Toronto', 'Vancouver'), loc='upper right')
plt.xlabel("Month")
plt.ylabel("Temperature")
plt.title("2012 Average Temperature (100558261)", fontsize=12)
plt.axis([0,13,-10,25])

#Shows the plot
plt.show()
```
