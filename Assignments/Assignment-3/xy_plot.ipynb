```python
#Jesse McGough
#100558261

#Imports what is needed
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

#Opens the file
point = open("xy_points.dat", "r")

#Defines arrays to hold the data
x_list = []
y_list =[]

#Fills the arrays
for value in point.readline().split(' '):
    x_list.append(float(value))

for value in point.readline().split(' '):
    y_list.append(float(value))

#Turns the values into vectors
x = np.array(x_list)
y = np.array(y_list)

#Isolates the values between 0 and 5 on the y-axis and less than -2 on the x-axis
x_ineq = x[(x<-2)&(y>=-5)&(y<=0)]
y_ineq = y[(x<-2)&(y>=-5)&(y<=0)]

#Isolates the values within a 5 radius from (0,5)
x_circle = x[((((x)**2+(y-5)**2)**(0.5))<=5)]
y_circle = y[((((x)**2+(y-5)**2)**(0.5))<=5)]

#Creates the plot, with the required details
plt.figure(num=None, figsize=(10,10), dpi=100, facecolor='white', edgecolor='black')
plt.scatter(x, y, color='red')
plt.scatter(x_ineq, y_ineq, color='blue')
plt.scatter(x_circle, y_circle, color='green')
plt.axis([-10,10,-10,10])
plt.title("Highlighted Points (100558261)", fontsize=12)

#Shows the plot
plt.show()
```
